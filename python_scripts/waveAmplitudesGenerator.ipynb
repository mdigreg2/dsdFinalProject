```python
import numpy as np
import matplotlib.pyplot as pt
import pandas as pd
```

```python
domain = np.arange(0, 2*np.pi, (2*np.pi)/512)
len(domain)
```

```python
sin_amps = np.sin(domain)
sin_amps
```

```python
pt.plot(domain, sin_amps, linewidth=1.5, label="Sin Wave")
pt.legend()
pt.show()
```

```python
np.savetxt("sinwave_amplitudes.csv", sin_amps, delimiter=",")
```
